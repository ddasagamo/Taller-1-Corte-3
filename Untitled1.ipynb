```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

librerias = [
    'Casita', 'Librería Gadner', 'Librería Euler', 'Librería Konisberg',
    'Librería Gauss', 'Librería Voronoi', 'Librería Fahrenheit',
    'Librería Celsius', 'Librería Fibonacci', 'Librería Hilbert', 'Librería Richter'
]

distancias = {
    ('Casita', 'Librería Voronoi'): 190,
    ('Casita', 'Librería Euler'): 330,
    ('Casita', 'Librería Gauss'): 180,
    ('Casita', 'Librería Fahrenheit'): 180,
    ('Casita', 'Librería Fibonacci'): 345,
    ('Librería Voronoi', 'Librería Euler'): 317,
    ('Librería Voronoi', 'Librería Fahrenheit'): 190,
    ('Librería Euler', 'Librería Konisberg'): 300,
    ('Librería Gauss', 'Librería Konisberg'): 275,
    ('Librería Gauss', 'Librería Richter'): 198,
    ('Librería Gauss', 'Librería Fahrenheit'): 314,
    ('Librería Fibonacci', 'Librería Gadner'): 299,
    ('Librería Fibonacci', 'Librería Richter'): 450,
    ('Librería Fibonacci', 'Librería Hilbert'): 250,
    ('Librería Fahrenheit', 'Librería Celsius'): 255,
    ('Librería Fahrenheit', 'Librería Hilbert'): 230,
    ('Librería Hilbert', 'Librería Celsius'): 312,
    ('Librería Richter', 'Librería Konisberg'): 225,
    ('Librería Gadner', 'Librería Euler'): 415,
    ('Librería Gadner', 'Librería Voronoi'): 170,
    ('Librería Gadner', 'Librería Fibonacci'): 310
}


adj_matrix = np.zeros((len(librerias), len(librerias)), dtype=int)
for (lib1, lib2), distancia in distancias.items():
    i, j = librerias.index(lib1), librerias.index(lib2)
    adj_matrix[i][j] = distancia
    adj_matrix[j][i] = distancia

print("Matriz de Adyacencia:")
print(adj_matrix)

inc_matrix = np.zeros((len(librerias), len(distancias)), dtype=int)
for idx, ((lib1, lib2), distancia) in enumerate(distancias.items()):
    i, j = librerias.index(lib1), librerias.index(lib2)
    inc_matrix[i][idx] = 1
    inc_matrix[j][idx] = 1

print("\nMatriz de Incidencia:")
print(inc_matrix)


G = nx.Graph()
for (lib1, lib2), distancia in distancias.items():
    G.add_edge(lib1, lib2, weight=distancia)


camino_mas_corto = nx.shortest_path(G, source='Casita', target='Librería Celsius', weight='weight')
distancia_mas_corta = nx.shortest_path_length(G, source='Casita', target='Librería Celsius', weight='weight')
print("\nCamino más corto desde Casita hasta Librería Celsius:", camino_mas_corto)
print("Distancia más corta desde Casita hasta Librería Celsius:", distancia_mas_corta)


plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=10)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d for (u, v, d) in G.edges(data='weight')})
plt.show()
```
